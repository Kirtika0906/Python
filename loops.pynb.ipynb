```python
salary = int(input("Enter salary: "))
if salary<=10000:
    print("you've got bonus of 10%")
elif salary<=20000:
    print("you've got bonus of 5%")
elif salary<=30000:
    print("you've got bonus of 3%")
else:
    print("you've got bonus of 2%")
```
