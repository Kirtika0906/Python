```python
#sets
fruits={"apple","mango"}
fruits.add("banana")
print(fruits)
```

```python
fruits.remove("mango")
print(fruits)
```

```python
a={1,2,3,7,8}
b={3,4,5,6,7}
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(b.difference(a))
```
